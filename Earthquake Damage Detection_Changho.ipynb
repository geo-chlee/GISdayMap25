
<h1 style="text-align:center;"> Earthquake Damage Detection through Land-Use and Parcel-Based Analysis Using SAR-Derived Coherence Change</h1>
<center> <b> Changho Lee, The University of Texas at Dallas </b> </center>

> <b> Abstract </b>

> Earthquakes cause severe surface and economic damage, especially in urban areas, yet limited research integrates land parcel–based data with remotely sensed observations for damage assessment. Traditionally, remote sensing (RS) and socio-economic GIS analyses have been conducted separately—RS focuses on pixel-based surface changes, while GIS relies on survey-based damage records. This study bridges these two approaches by integrating synthetic aperture radar (SAR) interferometry with cadastral parcel data containing land-use information. Spatial statistical methods such as local Moran’s <i>I</i> and Lee’s <i>L</i> bivariate local spatial association are used. Focusing on the November 15, 2017, Pohang earthquake—the most economically destructive earthquake in South Korea—the results revealed distinct spatial regimes: natural areas such as forests and rice paddies dominated high–high clusters, while artificial surfaces like building sites and roads characterized low–low clusters. Bivariate analysis further identified high-loss-coherence/high-displacement zones as key indicators of earthquake-induced damage. This integrated approach demonstrates how combining differential interferometric SAR and spatial autocorrelation improves the detection of seismic impacts by enhancing urban resilience and guiding post-disaster spatial planning.

> **Note** <br>
> All GIS analyses were conducted using ArcGIS Pro, Pandas, numpy, and arcpy, while Sentinel-1 SAR data were processed with ESA SNAP.<br>

<h2 style="color: blue; text-align: left;">
  <b>Introduction</b>: Pohang earthquake on November 15, 2017
<a id="Introduction"></a>
</h2>

- **Date & Time:** November 15, 2017, at 2:29 PM (location time, KST)
- **Location:** Epicenter at 129.36°E, 36.12°N (Korea Meteorological Administration, KMA) as shown in <a href="#fig1">Figure 1</a>.
- **Magnitude:** M5.4, the second-largest earthquake in South Korea’s modern seismic record (KMA)
- **Economic impact:** Estimated damage of ≈731 million USD, the most financially devastating earthquake in Korean history (National Disaster Safety Portal, 2025)
- **Geological features** (Naik et al., 2019)
    - Heunghae Basin (epicenter) is underlain by recent alluvial sediments with poor subsoil conditions, leading to amplified seismic waves
    - Shear Wave Velocity: Ranged between 120–275 m/s, indicating high susceptibility to ground motion amplification.
    - Fault Type: Reverse-dominant oblique-slip fault, striking northeast–southwest, with a focal depth of 7 km
    - Ground Motion: Peak Ground Acceleration (PGA) of 0.58g, equivalent to VIII–IX intensity on the Modified Mercalli Intensity (MMI) scale

<h2 style="color: blue; text-align: left;">
  <b>Analysis I</b>: Synthetic aperture radar (SAR) and Korean cadastral parcel data in Pohang
    <a id="Analysis I"></a>
</h2>

- #### Synthetic aperture radar (SAR)
    - A remote sensing system that uses microwave radar to capture high-resolution images of the Earth’s surface.
    - When multiple SAR images of the same area are collected at different times, Interferometric SAR (InSAR) techniques can detect ground displacement.
    - Differential InSAR (DInSAR) compares pre- and post-event images to measure surface deformation and coherence change.
    - In the study, Sentinel-1 SAR imagery from the European Space Agency (ESA) is used and processed as following the procedures illustrated in <a href="#fig2">Figure 2</a>.
    - Temporal coverage
        - Coherence change was computed from six scenes across three time periods (10/23, 11/04, 11/16) surrounding the November 15, 2017 earthquake.
        - 64 scenes (May 8, 2017 – May 15, 2018) were used to create 16 independent interferogram pairs capturing pre- and post-event surface deformation.

- #### Korean Cadastral Parcel Data (Pohang Case Study)
    - Data serves as the official registry of land boundaries, ownership, and land-use types in South Korea.
    - Each polygon in the cadastral map is categorized by one of 28 land-use types (with three categories: Artificial, Natural, and Mixed use).
    - In Pohang, 74,356 parcels existed in the 2017 dataset; 69,515 parcels were used after excluding areas smaller than one SAR pixel (14 m × 14 m = 196 m²).
    - In this study, cadastral parcels are used for the spatial analysis units by aggregating line-of-sight (LOS) displacement and coherence values.
    - <a href="#fig3">Figure 3</a>, <a href="#tab12">Table 1</a>, and <a href="#tab12">Table 2</a> show the distributions of cadastral parcels according to land use types and categories.
- #### Advantages of combining SAR and Parcel data
    - Linking physical surface changes with land ownership and socio-economic meaning.
    - Providing a consistent, noise-reduced value than pixels and aligning with legal and administrative boundaries.

<h2 style="color: blue; text-align: left;">
  <b>Analysis II</b>: Coherence change between pre- and post-earthquake and surface deformation
    <a id="Analysis II"></a>
</h2>

- **Coherence:** measures the similarity or correlation between two SAR images acquired over the same area at different times (Touzi et al. 1999).
    - **Value range:** from 0 to 1, where 1 indicates perfect similarity (no change in surface scattering properties), and 0 indicates complete decorrelation (significant change or noise).
- **Coherence change:** Difference between pre- and post-earthquake phase coherence to detect damaged areas by the earthquake (as shown in <a href="#fig4">Figure 4</a>).
    - **Equation:** <i>Coh</i><sub>diff</sub> = <i>Coh</i><sub>post</sub> &minus; <i>Coh</i><sub>pre</sub>, where positive and negative values indicate loss and gain respectively.
    - **Loss in coherence:** indicates increased surface disruption, such as structural damage, soil displacement, or vegetation alteration
    - **Near 0 in coherence change:** represents stable surfaces with minimal physical change.
    - **Gain in coherence:** indicates areas probably unaffected by the earthquake, reflecting a change from previously unstable or noisy surfaces to more stable and consistent surface after the event.
    - Note: Detailed results are illustrated in Figures <a href="#fig5">5</a>–<a href="#fig7">7</a>.
- **Surface deformation:** In DinSAR, surface deformation refers to the movement of the ground surface between two radar acquisitions. This movement is measured along the radar’s line of sight (LOS) — the direct path between the satellite sensor and the ground target.
    - **Positive value:** indicates the surface movement is away from the satellite.
    - **Negative value:** indicates the surface movement is toward the satellite.
    - Note: Detailed results are shown in <a href="#fig8">Figure 8</a>.

<a id="fig4"></a>
<center><img src="fig4_coherencechangemap.png" alt="Figure 4" width="90%"></center>
  <br><span style="font-size: 0.95em;"><b>Figure 4.</b> Spatial distribution of coherence change in the Pohang study area.</span>

> **Map description:** Coherence change was calculated as the difference between pre-earthquake coherence (October 23–November 4, 2017) and post-earthquake coherence (November 4–16, 2017). Positive values indicate coherence gain, representing a change from decorrelation to correlation, whereas negative values indicate coherence loss (i.e., decorrelation) after the earthquake.
>
> **Map interpretation:** Overall, coherence change values are predominantly negative in Pohang. Distinct zones of negative coherence change—indicating coherence loss—are concentrated in the southern to southeastern portions of the study area, while the northern areas are dominated by coherence gain. Coherence change also varies by land-use type: artificial and mixed land-use categories generally exhibit decreased coherence, whereas natural land-use areas show more variable patterns.

<h2 style="color: blue; text-align: left;">
  <b>Analysis III</b>: Identifying spatial patterns using Local Moran's <i>i</i> and Local Lee's <i>L</i>
    <a id="Analysis III"></a>
</h2>

- **Local Moran's <i>i</i>:** univariate local spatial autocorrelation method (Anselin, 1995), measures whether a feature and its surrounding neighbors exhibit significantly higher or lower values than expected under spatial randomness. Detailed description of Local Moran's <i>i</i> is available at <a href="https://pro.arcgis.com/en/pro-app/latest/tool-reference/spatial-statistics/h-how-cluster-and-outlier-analysis-anselin-local-m.htm" target="_blank">ArcGIS Pro: How Cluster and Outlier Analysis (Anselin Local Moran's I) works</a>
    - In the study, local Moran's <i>i</i> is used to identify clusters of coherence change (<a href="#fig9">Figure 9</a>) and clusters of surface displacement (<a href="#fig11">Figure 11</a>).
<br><br>

- **Local Lee's <i>L</i>:** measures the spatial dependence between two continuous variables based on the correlation coefficient by evaluating the similarity of their spatial patterns at the local level (Lee, 2001). Detailed description of Lee's <i>L</i> is available at <a href="https://pro.arcgis.com/en/pro-app/latest/tool-reference/spatial-statistics/how-bivariate-spatial-association-works.htm" target="_blank">ArcGIS: Bivariate Spatial Association (Lee's L) (Spatial Statistics)</a>
    - In the study, local Lee's <i>L</i> statistic was applied to identify spatial clusters that represent the combined patterns of coherence change and surface displacement (<a href="#fig13">Figure 13</a>). 

<a id="fig9"></a>
<center><img src="fig9_cohCluster.png" alt="Figure 9" width="70%">
  <br><span style="font-size: 0.95em;"><b>Figure 9.</b> Spatial regime map based on coherence change (<i>Coh</i><sub>diff</sub> = <i>Coh</i><sub>post</sub>-<i>Coh</i><sub>pre</sub>).</span></center>

> **Results:** This study focuses on the two cluster types. A High-High (HH) cluster represents parcels with coherence change values higher than the mean (−0.03m), surrounded by six neighboring parcels that also exceed the mean. Conversely, a Low-Low (LL) cluster indicates parcels and their neighbors with values lower than the mean. Reflecting the spatial pattern of coherence change, HH clusters are primarily concentrated in the central to northern portions of the study area, whereas LL clusters dominate the southern to eastern parts.

<a id="fig10"></a>
<center><img src="fig10_cohClusterPlot.png" alt="Figure 10" width="70%">
  <br><span style="font-size: 0.95em;"><b>Figure 10.</b> Land-use area percentages by coherence change cluster type.</span></center>

> **Results:** The dominant land-use categories account for more than 10% of each cluster.
> - **Overall trend:** HH (High–High) clusters are dominated by natural land uses, while LL (Low–Low) clusters are mainly artificial.<br>
> - **HH clusters:** consist of predominantly rice paddies (47.6%) and forests (32.3%), with smaller shares of farmland (6.0%), drainage (4.0%), and roads (3.6%).<br>
> - **LL clusters:** Largely compose of building sites (31.3%), forests (19.5%), roads (14.9%), and rice paddies (13.3%), along with smaller portions of farmland (5.5%), factory sites (2.5%), and mixed uses such as school sites (3.0%) and miscellaneous land (2.3%).<br>
> - **Other land uses:** Aggregated under “other,” showing minor or inconsistent proportions across both clusters.<br>

<center><img src="fig11_dispCluster.png" alt="Figure 11" width="70%">
  <br><span style="font-size: 0.95em;"><b>Figure 11.</b> Spatial regime map based on absolute value of displacement. </span></center>

> **Results:** There are two dominant displacement patterns corresponding to displacement magnitude: LL clusters, representing parcels with values lower than the mean displacement (0.027 m) and shown in blue, and HH clusters, representing parcels with values higher than the mean and shown in red. The spatial configuration of these clusters indicates that HH clusters are primarily concentrated in the western portion of the study area, while LL clusters are distributed in the eastern part, forming peripheral zones around the HH clusters.

<a id="fig12"></a>
<center><img src="fig12_dispClusterPlot.png" alt="Figure 12" width="70%">
  <br><span style="font-size: 0.95em;"><b>Figure 12.</b> Land-use area percentages by absolute displacement cluster type. <br> </center>
      <b>Note :</b> The study area columns (green bars) serve as a reference, indicating that there is no distinct relationship between displacement clusters and land-use types, as evidenced by the similar bar lengths across cluster types.</span>

> **Results:**
> - Overall pattern
>     - The land-use types of HH and LL clusters suggest that earthquake-induced displacement is not significantly related to their land-use types.
> - Cluster differences
>     - Artificial land uses (e.g., building sites and roads) occupy larger portions in LL clusters — 8.0% and 6.2%, respectively — compared to 3.2% and 2.5% in HH clusters.
>     - Natural land uses, particularly forests (58.9%) and rice paddies (20.6%), are more prevalent in HH clusters than in LL clusters (52.5% and 16.3%, respectively).
>     - Mixed-use categories such as school sites (0.8–0.4%) and miscellaneous land (1.1–0.6%) remain minor components in both clusters.

<a id="fig13"></a>
<center><img src="fig13_leeL.png" alt="Figure 13" width="70%"></center>
  <br><span style="font-size: 0.95em;"><b>Figure 13.</b> Bivariate local association (Lee’s <i>L</i>) spatial pattern map. Note that low coherence indicates a decrease in correlation. L-H areas represent positive correlation, which means positive coherence change (drop in coherence between post- and pre-events) as displacement increases.</span>

> **Results:** This study specifically focuses on Low coherence change – High displacement (L–H) areas, which are sparsely distributed in the central, western, and northwestern parts of the study area. These zones are considered the most susceptible to earthquake-induced damage due to the combined effects of surface deformation and coherence loss.
> 
> - **Dominant land-use type in L-H**
>     - L–H areas are largely associated with natural land uses, accounting for 82.1% of the total cluster area.
>     - Forests (51.6%) and rice paddies (17.8%) dominate this category, showing moderate coherence losses of −0.129 and −0.167, respectively.
> - **Artificial land uses**
>     - Represent 15.7% of the total, mainly building sites (7.7%) and roads (4.4%), both showing stronger coherence losses (about −0.17 to −0.18).
> - **Mixed-use areas**
>     - Account for 2.2%, with coherence loss (−0.158) similar to the overall mean (−0.163).
> - **Interpretation**
>     - L–H clusters are concentrated in vegetated and cultivated areas, where coherence remains moderately stable despite strong displacement.
>     - Artificial surfaces experience greater decorrelation, indicating more significant radar signal disruption under similar ground-motion conditions.

<h2 style="color: blue; text-align: left;">
  <b>Discussion and Conclusion</b>
    <a id="Discussion"></a>
</h2>


- **Methodological contribution**: Integrated SAR interferometry with cadastral parcel data using local Moran’s <i>i</i> and Lee’s <i>L</i> to link physical deformation with land-use context based on parcel as a spatial unit.
- **Spatial findings**: In univariate local Moran's <i>i</i> results of coherence change,
    - **HH clusters** were mainly associated with natural surfaces, particularly forests and rice paddies. The rice paddies exhibited temporarily higher coherence after harvest, as vegetation removal and exposure of bare soil increased backscattering stability (<a href="#fig14">Figure 14</a>).
    - **LL clusters** were concentrated on artificial surfaces, including building sites and roads, which showed greater coherence loss linked to surface disturbance and structural instability after the earthquake.
- **Key insight**: In bivariate spatial association statistics, Lee's <i>L</i> results, L–H clusters revealed critical earthquake damage zones where coherence dropped but displacement increased. Notably, one of these clusters—the downtown area of Heunghae-eup—corresponded to the most socio-economic damaged region (<a href="#fig15">Figure 15</a>).
- **Implication**: Parcel-level spatial analysis strengthens interpretation of seismic impacts and supports urban resilience and recovery planning.
- **Future work**: This study can be extended to economic damage evaluation by incorporating land value information from the cadastral parcel data (<a href="#fig16">Figure 16</a>).
